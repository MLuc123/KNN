````python
from keras.datasets import mnist
df = mnist.load_data('minst.db')
train,test = df
X_train, y_train = train
X_test, y_test = test
````

````python
import matplotlib.pyplot as plt
import numpy as np
np.set_printoptions(edgeitems=30, linewidth=100000)
for i in range(5):
    print(y_test[i],'\n')
    print(X_test[i],'\n')
    plt.contourf(np.rot90(X_test[i].transpose()))
    plt.show()
````

````python
(X_train, y_train), (X_test, y_test) = mnist.load_data()

print(f"X_train shape: {X_train.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"Single image shape: {X_train[0].shape}")
````

````python
X_train_final = X_train.reshape(60000, 784)
X_test_final = X_test.reshape(10000, 784)

X_train_final = X_train_final / 255.0
X_test_final = X_test_final / 255.0
````

###Optimal Neighbors
For the dataset, the optimal number of neighbors is around k=3 or k=5.

k=1 is often too sensitive to noise (a single weirdly written digit can throw off the prediction).

k > 10 starts to smooth out the decision boundaries too much, causing  digits to blur together.

### Accuracy Report
With k=3,the accuracy is approximately 94%. This means out of 10,000 test images, the model correctly identifies about 9,400 of them.

### Matrix Analysis
The confusion matrix shows that the model is highly accurate along the diagonal (correct predictions). However, there are clusters of off-diagonal values indicating specific, but consistent errors.


````python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train_final[:10000], y_train[:10000])

y_pred = knn.predict(X_test_final)
````

````python
print(f"Accuracy: {accuracy_score(y_test, y_pred) * 100:.2f}%")

cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap='Blues')
plt.title("Confusion Matrix")
plt.show()
````

### Most likely errors
4 vs 9:  A handwritten 4 often looks like a 9 if the top loop is closed.

7 vs 1: A 7 often looks like a slanted 1.

3 vs 8: Both digits are loopy and round; messy handwriting makes look similar.

### Patterns
The errors are geometric. The computer makes mathematical mistakes based on pixel overlap. It rarely confuses distinct shapes like 0 and 1 because their black pixels are in completely different locations in the matrix.

### Color Vision
For grayscale, there was just a single 2D matrix height x width because we only needed one number per pixel to show brightness.

To represent a color image, we  need three matrices stacked on top of each other, one for Red, one for Green, and one for Blue, making the shape (28, 28, 3).

To make this work with a model like kNN we have to flatten all three colors.

**Here is how I would do it:**

Take the Red 2828 matrix and make it a vector of 784 numbers and do the same for the Green and blue.Then, join them all into one giant row, making it now 3x as long.

Basically, the first part of the row tells the computer how red each pixel is, the middle part tells it how green, and the end tells it how blue. The computer doesn't really see the colors; it just sees a row of numbers that is three times longer than the black-and-white version.
